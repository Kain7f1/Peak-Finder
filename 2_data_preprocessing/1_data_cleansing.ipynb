# 1. 데이터 정제 data cleansing

## 1-a. 우리 프로젝트의 목적

- 투자자의 “감정”을 읽는 것이 주 목적이다.
- 그러므로 숫자와 대부분의 특수문자는 무의미하다
- 한글 감정 표현을 잡아내는 것이 주 목적이다

## 1-b. 정제 계획

- 특수 :  - ! ? 를 제외하고 전부 공백 문자로 치환한다
 :    - 공백 문한다자로 치환 : 
    - 소문자로 변경한다

## 1-c. 정제 진행
1. 영어 소문자로 통일
2. . 한글, 영어, !? 만 남기고 전부 공백문자로 치환한다3. 2. 공백 문자가 연속적으로 있는 걸 1개로 줄4. 
3. 연속되는 !, ? 를 최대 1개로5. 
4. 연속되는 한글 자음, 모음을 최대 3개로 줄임줄인다

### 1-c-1. 데이터 불러오기

```python
# 데이터 읽어오기
import pandas as pd
df = pd.read_csv("1_id_text.csv", encoding="utf-8")
df.tail(3)
```

### 1-c-2. 영어 소문자로 통일

```python
df['lowercase_text'] = df['text'].str.lower()
df.head(15)
```

### 1-c-3. 한글, 영어, !? 만 남기고 전부 공백문자로 치환한다.

```python
import re

# 한글, 영어, !, ? 만 남기고 나머지를 공백으로 치환하는 함수 정의
def replace_chars(text):
    return re.sub('[^a-z가-힣ㄱ-ㅎㅏ-ㅣ!?]', ' ', text)

# 데이터프레임의 'text' 칼럼에 함수 적용
df['filtered_text'] = df['lowercase_text'].apply(replace_chars)
df.head(30)
```

### 1-c-4. 연속되는 !, ? 를 최대 1개로 줄임

```python
# 연속되는 !, ? 를 최대 1개로 줄이는 함수 정의
def reduce_excl_quest_marks(text):
    text = re.sub('[!]+', '!', text)
    return re.sub('[?]+', '?', text)

# 데이터프레임의 'cleaned_text' 칼럼에 함수 적용
df['reduced_punct_text'] = df['filtered_text'].apply(reduce_excl_quest_marks)
df.head(30)
```

### 1-c-5. 연속되는 한글 자음, 모음 최대 3개로 줄이기

```python
from soynlp.normalizer import emoticon_normalize
def reduce_korean_text(input_text):
    normalized_text = emoticon_normalize(input_text, num_repeats=3)
    return normalized_text

df['reduced_korean_text'] = df['reduced_punct_text'].apply(reduce_korean_text)
df.head(30)
```

### 1-c-6. 연속적인 공백 문자 최대 1개로 줄임

```python
def reduce_spaces(text):
    # 연속되는 공백을 하나의 공백으로 치환
    return re.sub(' +', ' ', text)

df['cleansed_text'] = df['reduced_korean_text'].apply(reduce_spaces)
df.head(30)
```

### 1-c-7. 불필요한 칼럼을 제거한다

```python
# 제거할 칼럼 목록 (예시: 'lowercase_text', 'cleaned_text')
columns_to_remove = ['lowercase_text', 'cleaned_text', 'reduced_punct_text', 'reduced_korean_text', 'text']

# 해당 칼럼들을 데이터프레임에서 제거
df_dropped = df.drop(columns=columns_to_remove)
df_dropped.head(30)
```

### 1-c-8. 파일로 저장한다

```python
# .csv 파일로 저장한다
df_dropped.to_csv("1_cleansed_text.csv", index=False, encoding="utf-8")
```

## 1-d. 함수화
입력 - .csv 파일의 경로만 지정하면 모든 과정을 수행할 수 있도록 함수화한다.

### 1-d-1. 함수 만들기

```python
# 시간 측정 데코레이터
import time

def time_decorator(func):
    def wrapper(*args, **kwargs):
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f"Function '{func.__name__}' executed in {end_time - start_time} seconds.")
        return result
    return wrapper
```

```python
import pandas as pd
from soynlp.normalizer import emoticon_normalize
    
@time_decorator
def make_cleansed_csv_file(read_file_path, text_column_name, save_file_path):
    
    # 1. 데이터 읽어오기
    print("[progress 1/8] 데이터 읽어오기")
    df = pd.read_csv(read_file_path, encoding="utf-8")

    # 2. 영어는 소문자로 변경
    print("[progress 2/8] 영어는 소문자로 변경")
    df['lowercase_text'] = df[text_column_name].str.lower()

    # 3. 한글, 영어, !? 만 남기고 전부 공백문자로 치환
    print("[progress 3/8] 한글, 영어, !? 만 남기고 전부 공백문자로 치환")
    def replace_chars(text):
        return re.sub('[^a-z가-힣ㄱ-ㅎㅏ-ㅣ!?]', ' ', text)
    df['filtered_text'] = df['lowercase_text'].apply(replace_chars)

    # 4. 연속되는 !, ? 를 최대 1개로 줄임
    print("[progress 4/8] 연속되는 !, ? 를 최대 1개로 줄임")
    def reduce_excl_quest_marks(text):
        text = re.sub('[!]+', '!', text)
        return re.sub('[?]+', '?', text)
    df['reduced_punct_text'] = df['filtered_text'].apply(reduce_excl_quest_marks)

    # 5. 연속되는 한글 자음, 모음 최대 3개로 줄이기
    print("[progress 5/8] 연속되는 한글 자음, 모음 최대 3개로 줄이기")
    def reduce_korean_text(input_text):
        normalized_text = emoticon_normalize(input_text, num_repeats=3)
        return normalized_text
    df['reduced_korean_text'] = df['reduced_punct_text'].apply(reduce_korean_text)

    # 6. 연속적인 공백 문자 최대 1개로 줄임
    print("[progress 6/8] 연속적인 공백 문자 최대 1개로 줄임")
    def reduce_spaces(text):
        # 연속되는 공백을 하나의 공백으로 치환
        return re.sub(' +', ' ', text)
    df['cleansed_text'] = df['reduced_korean_text'].apply(reduce_spaces)

    # 7. 불필요한 칼럼을 제거
    print("[progress 7/8] 불필요한 칼럼을 제거")
    columns_to_remove = [text_column_name, 'lowercase_text', 'filtered_text', 'reduced_punct_text', 'reduced_korean_text']
    df_dropped = df.drop(columns=columns_to_remove)
    df.head(3)
    
    # 8. 파일로 저장
    print("[progress 8/8] 파일로 저장")
    df_dropped.to_csv(save_file_path, index=False, encoding="utf-8")
```

### 1-d-2. 함수 테스트

```python
read_file_path = "./1_id_text.csv"
text_column_name = "text"
save_file_path = "./1_cleansed_text.csv"
make_cleansed_csv_file(read_file_path, text_column_name, save_file_path)
```
