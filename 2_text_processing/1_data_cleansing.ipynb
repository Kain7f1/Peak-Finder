## 1-d. 부동산 갤러리에 해당되는 row 제거

## 1-e. 영어 소문자로 통일
- 결과 : df

```python
df.loc[:, 'lowercase_text'] = df['text'].str.lower()
df.head(15)
```

## 1-f. 한글, 영어, !? 만 남기고 전부 공백문자로 치환한다.
- 결과 : df

```python
import re

# 한글, 영어, !, ? 만 남기고 나머지를 공백으로 치환하는 함수 정의
def replace_chars(text):
    return re.sub('[^a-z가-힣ㄱ-ㅎㅏ-ㅣ!?]', ' ', text)

# 데이터프레임의 'text' 칼럼에 함수 적용
df.loc[:, 'filtered_text'] = df['lowercase_text'].apply(replace_chars)
df.head(30)
```

```python
df.info()
```

## 1-g. 연속되는 !, ? 를 최대 1개로 줄임
- 결과 : df

```python
# 연속되는 !, ? 를 최대 1개로 줄이는 함수 정의
def reduce_excl_quest_marks(text):
    text = re.sub('[!]+', '!', text)
    return re.sub('[?]+', '?', text)

# 데이터프레임의 'cleaned_text' 칼럼에 함수 적용
df['reduced_punct_text'] = df['filtered_text'].apply(reduce_excl_quest_marks)
df.head(30)
```

```python
df.info()
```

## 1-h. 연속되는 한글 자음, 모음 최대 2개로 줄이기
- 결과 : df

```python
from soynlp.normalizer import emoticon_normalize
def reduce_korean_text(input_text):
    normalized_text = emoticon_normalize(input_text, num_repeats=2)
    return normalized_text

df.loc[:, 'reduced_korean_text'] = df['reduced_punct_text'].apply(reduce_korean_text)
df.head(30)
```

```python
df.info()
```

## 1-i. 연속적인 공백 문자 최대 1개로 줄임
- 결과 : df

```python
def reduce_spaces(text):
    # 연속되는 공백을 하나의 공백으로 치환
    return re.sub(' +', ' ', text)

df.loc[:, 'cleansed_text'] = df['reduced_korean_text'].apply(reduce_spaces)
df.head(30)
```

```python
df.info()
```

## 1-j. 불필요한 칼럼을 제거한다
- 결과 : df_dropped

```python
# 제거할 칼럼 목록 (예시: 'lowercase_text', 'cleaned_text')
columns_to_remove = ['lowercase_text', 'filtered_text', 'reduced_punct_text', 'reduced_korean_text', 'text']

# 해당 칼럼들을 데이터프레임에서 제거
df_dropped = df.drop(columns=columns_to_remove)
df_dropped.head(30)
```

## 1-k. 파일로 저장한다

```python
# .csv 파일로 저장한다
df_dropped.to_csv("1_cleansed_data.csv", index=False, encoding="utf-8")
```
